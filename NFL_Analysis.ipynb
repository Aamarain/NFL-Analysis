```python
scores_with_cover = scores.copy()
```

```python
scores_cleaned = scores_with_cover.dropna(subset=['team_favorite_id', 'spread_favorite', 'over_under_line'])
scores_cleaned.head(10)
```

```python
scores_cleaned.isnull().sum()
```

```python
scores_final = scores_cleaned.drop(columns=['weather_humidity', 'weather_detail'])
scores_final.head(20)
```

```python
def check_favorite_cover(row):

    if pd.isna(row['team_favorite_id']) or pd.isna(row['spread_favorite']):
        return None
        
    if row['team_favorite_id'] == row['team_home']:
        favorite_score = row['score_home']
        underdog_score = row['score_away']
        spread = -row['spread_favorite']
    else:
        favorite_score = row['score_away']
        underdog_score = row['score_home']
        spread = -row['spread_favorite']
        
    if favorite_score + spread > underdog_score:
        return "Yes"
    elif favorite_score + spread == underdog_score:
        return "Push"
    else:
        return "No"
```

```python
scores_final['favorite_covered'] = scores_with_cover.apply(check_favorite_cover, axis=1)
```

```python
scores_final.head(5)
```

```python
scores_final['favorite_covered'].value_counts()
```

```python
def analyze_underdog_win(row):
    favorite = row['team_favorite_id']
    
    # Skip games without a favorite listed
    if pd.isna(favorite):
        return pd.Series([None, None])
    
    # Identify underdog and scores
    if row['team_home'] == favorite:
        underdog = row['team_away']
        underdog_score = row['score_away']
        favorite_score = row['score_home']
    else:
        underdog = row['team_home']
        underdog_score = row['score_home']
        favorite_score = row['score_away']
    
    # Check if the underdog won outright
    if underdog_score > favorite_score:
        underdog_won = 'Yes'
        margin = underdog_score - favorite_score
    else:
        underdog_won = 'No'
        margin = 0
    
    return pd.Series([underdog_won, margin])

# Apply the function across all rows
scores_final[['underdog_won', 'underdog_margin']] = scores_final.apply(analyze_underdog_win, axis=1)

# Display the first few rows with the new columns
scores_final.head(20)
```

```python
scores_final['schedule_date'] = pd.to_datetime(scores_final['schedule_date'], errors='coerce')

# Initialize counters
underdog_win_count = 0
underdog_loss_count = 0
```

```python
for index, row in scores_final.iterrows():
    # Check if the date is in the 1980s
    if row['schedule_date'].year >= 1980 and row['schedule_date'].year <= 1989:
        # Check if underdog_won is 'Yes'
        if row['underdog_won'] == 'Yes':
            underdog_win_count += 1
        elif row['underdog_won'] == 'No':
            underdog_loss_count += 1

# Print the results
print(f"Underdog Wins in the 1980s: {underdog_win_count}")
print(f"Underdog Losses in the 1980s: {underdog_loss_count}")
```

```python
# Determine the winner and margin
def get_winner_and_margin(row):
    if row['score_home'] > row['score_away']:
        winner = row['team_home']
        margin = row['score_home'] - row['score_away']
    elif row['score_away'] > row['score_home']:
        winner = row['team_away']
        margin = row['score_away'] - row['score_home']
    else:
        winner = 'Tie'
        margin = 0
    return pd.Series([winner, margin])

scores_final[['winner', 'win_margin']] = scores_final.apply(get_winner_and_margin, axis=1)

# Display results
scores_final.head(10)
```

```python
raiders_wins = scores_final[scores_final['winner'].str.contains('Raiders', na=False)]
```

```python
min_margin = raiders_wins['win_margin'].min()
max_margin = raiders_wins['win_margin'].max()

# Calculate how often each margin occurred
margin_counts = raiders_wins['win_margin'].value_counts().sort_index()

# Display the results
print(f"Raiders Minimum Margin of Victory: {min_margin}")
print(f"Raiders Maximum Margin of Victory: {max_margin}")

print("\nFrequency of Each Margin of Victory for Raiders:")
print(margin_counts)
```
