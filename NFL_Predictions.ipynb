```python
df = scores.dropna(subset=['team_favorite_id', 'spread_favorite', 'over_under_line'])
df.head(10)
```

```python
scores_df = df.copy()
```

```python
def get_winner(row):
    if row['score_home'] > row['score_away']:
        return row['team_home']
    elif row['score_away'] > row['score_home']:
          return row['team_away']
    else:
        return 'Tie'
```

```python
scores_df['winner'] = df.apply(get_winner, axis=1)
scores_df[['schedule_date', 'team_home', 'score_home', 'team_away', 
           'score_away', 'spread_favorite', 'total_points', 'winner']]
```

```python
scores_df.isnull().sum()
```

```python
scores_final = scores_df.drop(['schedule_date', 'schedule_season', 'schedule_week', 'schedule_playoff',
                        'team_favorite_id', 'stadium', 'stadium_neutral', 'weather_temperature', 
                        'weather_wind_mph', 'weather_humidity', 'weather_detail'], axis=1)
```

```python
scores_final.isnull().sum()
```

```python
scores_final['winner'] = scores_final['winner'].replace({
    'Phoenix Cardinals': 'Arizona Cardinals',
    'St. Louis Cardinals': 'Arizona Cardinals',
    'Tennessee Oilers': 'Tennessee Titans',
    'Houston Oilers': 'Tennessee Titans',
    'Los Angeles Raiders': 'Las Vegas Raiders',
    'Oakland Raiders': 'Las Vegas Raiders',
    'Washington Commanders': 'Washington Redskins',
    'Washington Football Team': 'Washington Redskins'})
```

```python
scores_final = scores_final.reset_index(drop=True)
scores_final.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
features = scores_final[['team_home', 'team_away', 'spread_favorite', 'over_under_line', 'score_home', 'score_away']]
target = scores_final['winner']
```

```python
home_encoder = LabelEncoder()
scores_final['team_home_encoded'] = home_encoder.fit_transform(scores_final['team_home'])
```

```python
away_encoder = LabelEncoder()
scores_final['team_away_encoded'] = home_encoder.fit_transform(scores_final['team_away'])
```

```python
scores_final.drop(columns=['team_home', 'team_away'], inplace=True)
```

```python
scores_final.head()
```

```python
scale_cols = ['score_home', 'score_away', 'total_points', 'team_home_encoded', 'team_away_encoded', 'spread_favorite', 'over_under_line']
```

```python
scores_final[['spread_favorite', 'over_under_line']].head(10)
```

```python
scores_final['spread_favorite'] = scores_final['spread_favorite'].replace('PK', '0')
scores_final['spread_favorite'] = scores_final['spread_favorite'].replace('PK', '0')
```

```python
scores_final['over_under_line'] = pd.to_numeric(scores_final['over_under_line'], errors='coerce')
```

```python
scores_final = scores_final.dropna(subset=['over_under_line'])
scores_final.head(10)
```

```python
scaler = StandardScaler()

scores_final[scale_cols] = scaler.fit_transform(scores_final[scale_cols])

scores_final[scale_cols].head()
```

```python
feature_cols = ['team_home_encoded', 'total_points', 'team_away_encoded', 'spread_favorite', 'over_under_line', 'score_home', 'score_away']

# X (features), y (target)
X = scores_final[feature_cols]
y = scores_final['winner']  # Assuming 'winner' is still the team name

# Optional: encode the winner column if it isn't already
from sklearn.preprocessing import LabelEncoder
winner_encoder = LabelEncoder()
y_encoded = winner_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestClassifier(
    n_estimators=100,        # Number of trees
    max_depth=None,          # No depth limit
    random_state=42
)

# Train the model
rf_model.fit(X_train, y_train)
```

```python
# Predict the test set
y_pred = rf_model.predict(X_test)

# Evaluate
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Detailed report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=winner_encoder.classes_))

# Optional: Confusion Matrix
import seaborn as sns
import matplotlib.pyplot as plt

conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=winner_encoder.classes_,
            yticklabels=winner_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import pandas as pd

report = classification_report(y_test, y_pred, target_names=winner_encoder.classes_, output_dict=True)

report_df = pd.DataFrame(report).transpose()

report_df = report_df.drop(['accuracy', 'macro avg', 'weighted avg'])

plt.figure(figsize=(14, 8))
report_df['f1-score'].sort_values(ascending=False).plot(kind='bar')
plt.title('F1-Score by Team')
plt.ylabel('F1-Score')
plt.xlabel('Team')
plt.xticks(rotation=90)
plt.show()
```
